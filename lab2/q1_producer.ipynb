```python
import time
from kafka import KafkaProducer
import json

producer = KafkaProducer(bootstrap_servers="localhost:9092",value_serializer=lambda m: json.dumps(m).encode('utf8'))
i = 0
search_term = 'World cup'
topic_name = 'raw-tweets'
while True:
    for tweet in tweepy.Paginator(client.search_recent_tweets, query=query, 
                              tweet_fields=['created_at','lang','possibly_sensitive'],max_results=10).flatten(limit=1):
        print('Tweet:',tweet)
        print('Tweet lang:',tweet['lang'])
        print('Tweet possibly sensitive:',tweet['possibly_sensitive'])
        print('\n')
        dictionary = {
            "Tweet": str(tweet),
            "Tweet lang": str(tweet['lang']),
            "Tweet possibly sensitive": str(tweet['possibly_sensitive']),
        }
        json_object = json.dumps(dictionary, indent=3)
        producer.send(topic_name, json_object)
    i += 1
    time.sleep(1)
```
