```python
from textblob import TextBlob
```

```python
# Create a function to get the polarity
def getPolarity(tweet: str) -> float:
    return TextBlob(tweet).sentiment.polarity


def getSentiment(polarityValue: int) -> str:
    if polarityValue < 0:
        return 'Negative'
    else:
        return 'Positive'
```

```python
import time
from kafka import KafkaConsumer
from kafka import KafkaProducer
import json

filter_topic_name_pos = 'positive-tweets'
filter_topic_name_neg = 'negative-tweets'
producer = KafkaProducer(bootstrap_servers="localhost:9092",value_serializer=lambda m: json.dumps(m).encode('utf8'))
filter_topic_name_en = 'en-tweets'
consumer = KafkaConsumer(filter_topic_name_en, bootstrap_servers="localhost:9092",value_deserializer=lambda m: json.loads(m.decode('utf-8')))#json.dumps(m).decode('utf8'))
for message in consumer:
    #print("Received message : {} from topic : {}, partition: {}, offset: {}",message)#.format(message.value, message.topic, message.partition, message.offset))
    tweets = json.loads(json.dumps(message.value))
    print(tweets)
    json_object = json.dumps(json.loads(tweets), indent=3)

    if getSentiment(getPolarity(json.loads(tweets)['Tweet'])) == 'Positive':
        producer.send(filter_topic_name_pos, json_object)
    
    elif getSentiment(getPolarity(json.loads(tweets)['Tweet'])) == 'Negative':
        producer.send(filter_topic_name_neg, json_object)

    #time.sleep(1)
```
